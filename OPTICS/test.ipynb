```python
import numpy as np 
import matplotlib.pyplot as plt
from point import *
from optics import *
```

```python
def loadData(filename):
    fin = open(filename)
    data = []
    for line in fin:
        line_list = line.strip().split("\t")
        d1 = float(line_list[0])
        d2 = float(line_list[1])
        data.append([d1,d2])
    return np.array(data)
```

```python
data = loadData("000_tmp2")
data
```

```python
fig = plt.figure()
ax =fig.add_subplot(111)
ax.scatter(data[:,1],data[:,0],c="green",s=0.5)
plt.show()
```

```python
points = []
for p in data:
    lat = p[0]
    lont = p[1]
    point = Point(lat,lont)
    points.append(point)
```

```python
len(points)
```

```python
optics = Optics(points, 150, 60)
```

```python
optics.run()  
```

```python
clusters = optics.cluster(50)
```

```python
fig = plt.figure()
ax =fig.add_subplot(111)
for cluster in clusters:
    points = cluster.points
    data = []
    for point in points:
        lat = point.latitude
        lont = point.longitude
        data.append([lat,lont])
    data = np.array(data)
    ax.scatter(data[:,1],data[:,0],s=0.5)
plt.show()
```
